```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Učitavanje CSV-a, automatski trime leading space i parse_dates
df = pd.read_csv(
    '../data/historical_consumption.csv',
    skipinitialspace=True,        # uklanja leading razmake iz imena stupaca
    parse_dates=['Datum'],        # sada će Datum biti datetime
    dayfirst=True
)

# 2) Preimenuj potrosnja → Potrošnja
df.rename(columns={'potrosnja':'Potrošnja'}, inplace=True)

# 3) Brzi pogled na strukturu
display(df.head())
print(df.info())
print('\nSažetak statistika:\n', df['Potrošnja'].describe())

# 4) Provjera nedostajućih vrijednosti
print('\nMissing values po stupcu:\n', df.isna().sum())

# 5) Linijski graf potrošnje kroz vrijeme
df.set_index('Datum')['Potrošnja'].plot(
    figsize=(12, 4), title='Povijesna potrošnja')
plt.xlabel('Datum')
plt.ylabel('Potrošnja')
plt.tight_layout()
plt.show()
```
